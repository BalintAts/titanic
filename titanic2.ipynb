```python
# OH encoding for sex
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

OH_encoder = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
OH_cols_Full_embarked = pd.DataFrame(OH_encoder.fit_transform(X_full[['Embarked']]))

#put back the index
OH_cols_Full_embarked.index = X_full.index



X_smaller = X_full.drop('Embarked', axis = 1)

#L_X_full = X_full.copy()
L_encoder = LabelEncoder()
X_smaller.Sex = L_encoder.fit_transform(X_full.Sex)
```

```python
X_numeric = X_smaller.drop(['Ticket'], axis = 1)
```

```python
#add back the embarked hots
X_numeric = pd.concat([X_numeric,OH_cols_Full_embarked], axis = 1)
```

```python
X_numeric.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_numeric, y,
                                                      train_size=0.8, test_size=0.2,
                                                      random_state=0)
print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
print(y_valid.shape)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, max_depth = 1
                       ,objective= 'reg:squarederror')
my_model.fit(X_train, y_train, early_stopping_rounds= 5,
            eval_set = [(X_valid, y_valid)], verbose = False)
```

```python
from sklearn.metrics import mean_absolute_error

predictions = my_model.predict(X_valid)
print(mean_absolute_error(predictions, y_valid))
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_valid, predictions.round(), normalize=True, sample_weight=None))
```
